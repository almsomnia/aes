```python
import pandas as pd
import re
import html

def clean_text_transformer(text):
   # Safety check
   if not isinstance(text, str):
      return ""
   
   # Decode HTML entities (&amp; -> &)
   text = html.unescape(text)

   # Remove non-alphanumeric chars (keep punctuations & spaces)
   text = re.sub(r'[^\w\s.,?!:;"\'\-]', '', text)

   # Replace newlines/tabs with regular space
   text = text.replace("\n", " ").replace("\t", " ").replace("\r", " ")

   # Whitespace normalization ("  I am   eating  " becomes "I am eating")
   text = re.sub(r'\s+', ' ', text).strip()

   return text
```

```python
DIR_DATA_A = "./dataset/data_A"

df_A_dev = pd.read_csv("{}/A_dev.csv".format(DIR_DATA_A))
df_A_test = pd.read_csv("{}/A_test.csv".format(DIR_DATA_A))
df_A_train = pd.read_csv("{}/A_train.csv".format(DIR_DATA_A))

DIR_DATA_B = "./dataset/data_B"

df_B_dev = pd.read_csv("{}/B_dev.csv".format(DIR_DATA_B))
df_B_test = pd.read_csv("{}/B_test.csv".format(DIR_DATA_B))
df_B_train = pd.read_csv("{}/B_train.csv".format(DIR_DATA_B))
```

```python
df_A_train["RESPONSE"] = df_A_train["RESPONSE"].apply(clean_text_transformer)
df_A_dev["RESPONSE"] = df_A_dev["RESPONSE"].apply(clean_text_transformer)
df_A_test["RESPONSE"] = df_A_test["RESPONSE"].apply(clean_text_transformer)

df_B_train["RESPONSE"] = df_B_train["RESPONSE"].apply(clean_text_transformer)
df_B_dev["RESPONSE"] = df_B_dev["RESPONSE"].apply(clean_text_transformer)
df_B_test["RESPONSE"] = df_B_test["RESPONSE"].apply(clean_text_transformer)
```

```python
df_A_train.to_csv("{}/.ignore/preprocessed/A_train.csv".format(DIR_DATA_A))
df_A_dev.to_csv("{}/.ignore/preprocessed/A_dev.csv".format(DIR_DATA_A))
df_A_test.to_csv("{}/.ignore/preprocessed/A_test.csv".format(DIR_DATA_A))

df_B_train.to_csv("{}/.ignore/preprocessed/B_train.csv".format(DIR_DATA_B))
df_B_dev.to_csv("{}/.ignore/preprocessed/B_dev.csv".format(DIR_DATA_B))
df_B_test.to_csv("{}/.ignore/preprocessed/B_test.csv".format(DIR_DATA_B))
```
