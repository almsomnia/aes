```python
import pandas as pd
import stanza
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
from sbert_feat_engineering import sbert_feat_engineering
from lang_feat_engineering import lang_feat_engineering

# sqdm init for pandas
tqdm.pandas()

DIR_DATA_A = "./dataset/data_A"
DIR_DATA_B = "./dataset/data_B"
```

```python
print("Loading model...")
# sbert_model = SentenceTransformer("indobenchmark/indosbert-base-p1")
# sbert_model = SentenceTransformer("firqaaa/indo-sentence-bert-base")
sbert_model = SentenceTransformer("paraphrase-multilingual-mpnet-base-v2")
stanza_pipeline = stanza.Pipeline("id", processors="tokenize,pos,lemma")

print("Model loaded")
```

```python
df_A_train = pd.read_csv("{}/.ignore/preprocessed/A_train.csv".format(DIR_DATA_A), keep_default_na=False)
df_A_dev = pd.read_csv("{}/.ignore/preprocessed/A_dev.csv".format(DIR_DATA_A), keep_default_na=False)
df_A_test = pd.read_csv("{}/.ignore/preprocessed/A_test.csv".format(DIR_DATA_A), keep_default_na=False)

df_B_train = pd.read_csv("{}/.ignore/preprocessed/B_train.csv".format(DIR_DATA_B), keep_default_na=False)
df_B_dev = pd.read_csv("{}/.ignore/preprocessed/B_dev.csv".format(DIR_DATA_B), keep_default_na=False)
df_B_test = pd.read_csv("{}/.ignore/preprocessed/B_test.csv".format(DIR_DATA_B), keep_default_na=False)
```

### Relevance Feature Extraction

```python
df_A_train_sbert = sbert_feat_engineering(
   df=df_A_train,
   stimulus_path="{}/stimulus.txt".format(DIR_DATA_A),
   model=sbert_model,
   save_path=None
)

df_A_dev_sbert = sbert_feat_engineering(
   df=df_A_dev,
   stimulus_path="{}/stimulus.txt".format(DIR_DATA_A),
   model=sbert_model,
   save_path=None
)

df_A_test_sbert = sbert_feat_engineering(
   df=df_A_test,
   stimulus_path="{}/stimulus.txt".format(DIR_DATA_A),
   model=sbert_model,
   save_path=None
)

df_B_train_sbert = sbert_feat_engineering(
   df=df_B_train,
   stimulus_path="{}/stimulus.txt".format(DIR_DATA_B),
   model=sbert_model,
   save_path=None
)

df_B_dev_sbert = sbert_feat_engineering(
   df=df_B_dev,
   stimulus_path="{}/stimulus.txt".format(DIR_DATA_B),
   model=sbert_model,
   save_path=None
)

df_B_test_sbert = sbert_feat_engineering(
   df=df_B_test,
   stimulus_path="{}/stimulus.txt".format(DIR_DATA_B),
   model=sbert_model,
   save_path=None
)
```

### Language Feature Extraction

```python
df_A_train_lang_only = df_A_train_sbert["RESPONSE"].progress_apply(lang_feat_engineering, pipeline=stanza_pipeline)
df_A_train_final = pd.concat([df_A_train_sbert, df_A_train_lang_only], axis=1)

df_A_dev_lang_only = df_A_dev_sbert["RESPONSE"].progress_apply(lang_feat_engineering, pipeline=stanza_pipeline)
df_A_dev_final = pd.concat([df_A_dev_sbert, df_A_dev_lang_only], axis=1)

df_A_test_lang_only = df_A_test_sbert["RESPONSE"].progress_apply(lang_feat_engineering, pipeline=stanza_pipeline)
df_A_test_final = pd.concat([df_A_test_sbert, df_A_test_lang_only], axis=1)

df_B_train_lang_only = df_B_train_sbert["RESPONSE"].progress_apply(lang_feat_engineering, pipeline=stanza_pipeline)
df_B_train_final = pd.concat([df_B_train_sbert, df_B_train_lang_only], axis=1)

df_B_dev_lang_only = df_B_dev_sbert["RESPONSE"].progress_apply(lang_feat_engineering, pipeline=stanza_pipeline)
df_B_dev_final = pd.concat([df_B_dev_sbert, df_B_dev_lang_only], axis=1)

df_B_test_lang_only = df_B_test_sbert["RESPONSE"].progress_apply(lang_feat_engineering, pipeline=stanza_pipeline)
df_B_test_final = pd.concat([df_B_test_sbert, df_B_test_lang_only], axis=1)
```

```python
from spelling_feat_engineering import check_spelling, load_dict

dictionary = load_dict('./supporting_data/kata-dasar.txt')

df_A_train_final["SPELLING_FEAT"] = df_A_train_final["RESPONSE"].progress_apply(lambda x: check_spelling(x, dictionary))
df_A_dev_final["SPELLING_FEAT"] = df_A_dev_final["RESPONSE"].progress_apply(lambda x: check_spelling(x, dictionary))
df_A_test_final["SPELLING_FEAT"] = df_A_test_final["RESPONSE"].progress_apply(lambda x: check_spelling(x, dictionary))

df_B_train_final["SPELLING_FEAT"] = df_B_train_final["RESPONSE"].progress_apply(lambda x: check_spelling(x, dictionary))
df_B_dev_final["SPELLING_FEAT"] = df_B_dev_final["RESPONSE"].progress_apply(lambda x: check_spelling(x, dictionary))
df_B_test_final["SPELLING_FEAT"] = df_B_test_final["RESPONSE"].progress_apply(lambda x: check_spelling(x, dictionary))
```

```python
df_A_train_final.to_csv("{}/.ignore/feature-extracted/A_df_train_final.csv".format(DIR_DATA_A))
df_A_dev_final.to_csv("{}/.ignore/feature-extracted/A_df_dev_final.csv".format(DIR_DATA_A))
df_A_test_final.to_csv("{}/.ignore/feature-extracted/A_df_test_final.csv".format(DIR_DATA_A))

df_B_train_final.to_csv("{}/.ignore/feature-extracted/B_df_train_final.csv".format(DIR_DATA_B))
df_B_dev_final.to_csv("{}/.ignore/feature-extracted/B_df_dev_final.csv".format(DIR_DATA_B))
df_B_test_final.to_csv("{}/.ignore/feature-extracted/B_df_test_final.csv".format(DIR_DATA_B))

df_train_final = pd.concat([df_A_train_final, df_B_train_final], ignore_index=True)
df_dev_final = pd.concat([df_A_dev_final, df_B_dev_final], ignore_index=True)
df_test_final = pd.concat([df_A_test_final, df_B_test_final], ignore_index=True)

df_train_final.to_csv("./dataset/.ignore/df_train_final.csv")
df_dev_final.to_csv("./dataset/.ignore/df_dev_final.csv")
df_test_final.to_csv("./dataset/.ignore/df_test_final.csv")
```
